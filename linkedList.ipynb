### Sort Binary LinkedList

```python
Problem Description

Given a Linked List A consisting of N nodes.

The Linked List is binary i.e data values in the linked list nodes consist of only 0's and 1's.

You need to sort the linked list and return the new linked list.

NOTE:

Try to do it in constant space.


Problem Constraints
1 <= N <= 105

A.val = 0 or A.val = 1



Input Format
First and only argument is the head pointer of the linkedlist A.



Output Format
Return the head pointer of the new sorted linked list.



Example Input
Input 1:

 1 -> 0 -> 0 -> 1
Input 2:

 0 -> 0 -> 1 -> 1 -> 0


Example Output
Output 1:

 0 -> 0 -> 1 -> 1
Output 2:

 0 -> 0 -> 0 -> 1 -> 1


Example Explanation
Explanation 1:

 The sorted linked list looks like:
  0 -> 0 -> 1 -> 1
Explanation 2:

 The sorted linked list looks like:
  0 -> 0 -> 0 -> 1 -> 1
```

```python
# Definition for singly-linked list.
# class ListNode:
#    def __init__(self, x):
#        self.val = x
#        self.next = None

class Solution:
    # @param A : head node of linked list
    # @return the head node in the linked list
    def solve(self, A):
        zeroC=0
        oneC=0
        num=A
        while num is not None:
            if num.val==0:
                zeroC+=1
            else:
                oneC+=1
            num=num.next
        #print(zeroC,oneC)
        if zeroC>0:
            nl=ListNode(0)
            zeroC-=1
        else:
            nl=ListNode(1)
            oneC-=1
        t=nl
        for i in range(zeroC):
            nn=ListNode(0)
            t.next=nn
            t=t.next
        for i in range(oneC):
            nn=ListNode(1)
            t.next=nn
            t=t.next
        return nl
```

## Reverse Link List II

```python
Reverse a linked list from position m to n. Do it in-place and in one-pass.

For example:
Given 1->2->3->4->5->NULL, m = 2 and n = 4,

return 1->4->3->2->5->NULL.

 Note:
Given m, n satisfy the following condition:
1 ≤ m ≤ n ≤ length of list. Note 2:
Usually the version often seen in the interviews is reversing the whole linked list
which is obviously an easier version of this question. 
```

```python
# Definition for singly-linked list.
# class ListNode:
#    def __init__(self, x):
#        self.val = x
#        self.next = None

class Solution:
    # @param A : head node of linked list
    # @param B : integer
    # @param C : integer
    # @return the head node in the linked list
    def reverseBetween(self, A, B, C):
        def reverseLinkedlist(head):
            pre=None
            current=head
            tail=head
            if not(current):
                return [head,head]
            nextNode=current.next
            
            while nextNode is not None:
                current.next=pre
                nextNext=nextNode.next
                nextNode.next=current
                pre=current
                current=nextNode
                nextNode=nextNext
            return [current,tail]
        prevOfStart=A
        ct=1
        while ct<(B-1):
            prevOfStart=prevOfStart.next
            ct+=1
        
        start=prevOfStart.next
        if B==1:
            start=A
        end=A
        ct=1
        while ct<C:
            end=end.next
            ct+=1
        nextOfEnd=end.next
        end.next=None
        
        reversedFromB=reverseLinkedlist(start)
        reversedFromBHead=reversedFromB[0]
        reversedFromBTail=reversedFromB[1]
        reversedFromBTail.next=nextOfEnd
        if B>1:
            prevOfStart.next=reversedFromBHead
        else:
            return reversedFromBHead
        return A
```

```python
arr=[int(x)for x in input().split()]
head=makeLinkedlistFromArr(arr)
q=Solution()
m=int(input("Enter m:"))
n=int(input("Enter n:"))
output=q.reverseBetween(head,m,n)
traverseLinkedlist(output)
```

## Remove Duplicates from Sorted List

```python
Given a sorted linked list, delete all duplicates such that each element appear only once.

For example,
Given 1->1->2, return 1->2.
Given 1->1->2->3->3, return 1->2->3.
```

```python
# Definition for singly-linked list.
# class ListNode:
#    def __init__(self, x):
#        self.val = x
#        self.next = None

class Solution:
    # @param A : head node of linked list
    # @return the head node in the linked list
    def deleteDuplicates(self, A):
        p=A
        nl=ListNode(p.val)
        np=nl
        prev=p.val
        p=p.next
        while p is not None:
            if p.val!=prev:
                nn=ListNode(p.val)
                np.next=nn
                np=np.next
                prev=p.val
            p=p.next
        return nl
```

## Merge Two Sorted LinkedList

```python
Merge two sorted linked lists and return it as a new list.
The new list should be made by splicing together the nodes of the first two lists, and should also be sorted.

For example, given following linked lists :

  5 -> 8 -> 20 
  4 -> 11 -> 15
The merged list should be :

4 -> 5 -> 8 -> 11 -> 15 -> 20
```

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    # @param A : head node of linked list
    # @param B : head node of linked list
    # @return the head node in the linked list
    def mergeTwoLists(self, A, B):
        newHeadVal=min(A.val,B.val)
        newHead=ListNode(newHeadVal)
        pA=A
        pB=B
        if pA.val==newHeadVal:
            pA=pA.next
        else:
            pB=pB.next
        p=newHead
        while ((pA is not None) and (pB is not None)):
            if pA.val<pB.val:
                valX=pA.val
                pA=pA.next
            else:
                valX=pB.val
                pB=pB.next
            nn=ListNode(valX)
            p.next=nn
            p=p.next
        while pA is not None:
            nn=ListNode(pA.val)
            p.next=nn
            p=p.next
            pA=pA.next
        while pB is not None:
            nn=ListNode(pB.val)
            p.next=nn
            p=p.next
            pB=pB.next
        return newHead
```

```python
class Node:
    def __init__(self,val):
        self.val=val
        self.next=None
def makeLinkedlistFromArr(arr):
    if len(arr)==0:
        return None
    head=Node(arr[0])
    
    head.next=makeLinkedlistFromArr(arr[1:])
    return head
def traverseLinkedlist(head):
    p=head
    while p is not None:
        print(str(p.val)+"->",end=" ")
        p=p.next
    print("None")
```

## Remove Nth node from list end

```python
Given a linked list, remove the nth node from the end of list and return its head.

For example,
Given linked list: 1->2->3->4->5, and n = 2.
After removing the second node from the end, the linked list becomes 1->2->3->5.

 Note:
If n is greater than the size of the list, remove the first node of the list.
Try doing it using constant additional space.
```

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

# Definition for singly-linked list.
# class ListNode:
#    def __init__(self, x):
#        self.val = x
#        self.next = None
class Solution:
    # @param A : head node of linked list
    # @param B : integer
    # @return the head node in the linked list
    def removeNthFromEnd(self, A, B):
        count=0
        p=A
        while p is not None:
            count+=1
            p=p.next
        if B>count:
            nodeToDel=0
        else:
            nodeToDel=count-B
        p=A
        if nodeToDel==0:
            return p.next
        c=0
        while c<(nodeToDel-1):
            p=p.next
            c+=1
        nextNext=p.next.next
        p.next=nextNext
        return A
```

## K reverse linked list

```python
Given a singly linked list and an integer K, reverses the nodes of the

list K at a time and returns modified linked list.

 NOTE : The length of the list is divisible by K 
Example :

Given linked list 1 -> 2 -> 3 -> 4 -> 5 -> 6 and K=2,

You should return 2 -> 1 -> 4 -> 3 -> 6 -> 5

Try to solve the problem using constant extra space.
```

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    # @param A : head node of linked list
    # @param B : integer
    # @return the head node in the linked list
    def reverseList(self, A, B):
        def reverseBetween(A, B, C):
            def reverseLinkedlist(head):
                pre=None
                current=head
                tail=head
                if not(current):
                    return [head,head]
                nextNode=current.next

                while nextNode is not None:
                    current.next=pre
                    nextNext=nextNode.next
                    nextNode.next=current
                    pre=current
                    current=nextNode
                    nextNode=nextNext
                return [current,tail]
            prevOfStart=A
            ct=1
            while ct<(B-1):
                prevOfStart=prevOfStart.next
                ct+=1

            start=prevOfStart.next
            if B==1:
                start=A
            end=A
            ct=1
            while ct<C:
                end=end.next
                ct+=1
            nextOfEnd=end.next
            end.next=None

            reversedFromB=reverseLinkedlist(start)
            reversedFromBHead=reversedFromB[0]
            reversedFromBTail=reversedFromB[1]
            reversedFromBTail.next=nextOfEnd
            if B>1:
                prevOfStart.next=reversedFromBHead
            else:
                return reversedFromBHead
            return A
        
        def lengOfLL(head):
            count=0
            while head is not None:
                count+=1
                head=head.next
            return count
        start=1
        end=start+B-1
        l=lengOfLL(A)
        op=A
        while end<=l:
            op=reverseBetween(op,start,end)
            start=end+1
            end=start+B-1
        return op
```

```python
class Node:
    def __init__(self,val):
        self.val=val
        self.next=None
def makeLinkedlistFromArr(arr):
    if len(arr)==0:
        return None
    head=Node(arr[0])
    
    head.next=makeLinkedlistFromArr(arr[1:])
    return head
def traverseLinkedlist(head):
    p=head
    while p is not None:
        print(str(p.val)+"->",end=" ")
        p=p.next
    print("None")
```

```python
arr=[int(x)for x in input().split(' ')]
A=makeLinkedlistFromArr(arr)
B=int(input())
q=Solution()
ans=q.reverseList(A,B)
traverseLinkedlist(ans)
```
